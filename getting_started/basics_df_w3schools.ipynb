```python
# Load pandas
import pandas as pd
```

```python
# Reading csv
# Getting mode of coloumn and filling NaN values
df = pd.read_csv('dirtydata.csv')
date_mode = df['Date'].mode()[0]
df['Date_2'] = df['Date'].fillna(date_mode,inplace=False)
df
```

```python
# Converting coloumn datatype
df = pd.read_csv('dirtydata.csv')
df['Date_2'] = pd.to_datetime(df['Date'])
df
```

```python
# Dropping NaN values
df = pd.read_csv('dirtydata.csv')
df.dropna(subset=['Date'],inplace=True)
df
```

```python
# Converting column value using apply function

df = pd.read_csv('dirtydata.csv')

df['Duration_updated'] = df['Duration'].apply(lambda x: 120 if x > 120 else x)

df
```

```python
# List of Tuples
matrix = [(22, 34, 23),
         (33, 31, 11),
         (44, 16, 21),
         (55, 32, 22),
         (66, 33, 27),
         (77, 35, 11)
         ]
# Create a DataFrame object
dfObj = pd.DataFrame(matrix, columns=list('xyz'), index=list('abcdef'))
dfObj['x'] = dfObj['x'].apply(lambda x : x**2)
dfObj
```
